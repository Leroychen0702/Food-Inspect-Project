```python
import pandas as pd

# Load CSV
df = pd.read_csv('Restaurant_and_Food_Establishment_Inspections__October_2016_to_January_2024__20250414.csv')

# Set display options
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_colwidth', None)  # Show full content in each cell
pd.set_option('display.width', None)  # Do not wrap rows

# Display first few rows with full column and row text
df.head()
```

```python
possible_missing_values = [
    'null', 'none', 'n/a', 'na', 'nan', 'missing', '0',
    '', ' ', '-', '--', 'n\\a', 'n.a.', 'n.a', 'na.', 'null.', 'none.',
    'no data', 'not available', 'not applicable', 'unavailable', 'tbd',
    'to be determined', 'nil', 'undefined', 'blank', 'empty', '?'
]

def is_any_missing(val):
    if pd.isna(val):
        return True
    if isinstance(val, str):
        return val.strip().lower() in possible_missing_values
    return False

missing_mask = df.applymap(is_any_missing)
missing_summary = missing_mask.sum()

# Show result
if (missing_summary > 0).any():
    print("⚠️ Columns with missing-like values:")
    print(missing_summary[missing_summary > 0])
else:
    print("✅ LOOKS GOOD — No missing, null, unknown, or blank values found.")
```

```python
import pandas as pd
import re
import swifter  # for parallel apply

# Function to extract code and description
def extract_code_and_description(text):
    if pd.isna(text):
        return pd.Series([None, None])
    
    # Match a number at the beginning, possibly followed by non-alphanumeric char (e.g., '*29 ')
    match = re.match(r'\*?(\d+)\s+(.*)', text.strip())
    if match:
        code = match.group(1)
        description = match.group(2).strip()
        return pd.Series([code, description])
    
    return pd.Series([None, text])

# Loop through Violation Description - 1 to 25
for i in range(1, 26):
    original_col = f'Violation Description - {i}'
    code_col = f'Violation Code - {i}'
    desc_col = f'Violation Description - {i}'

    if original_col in df.columns:
        # Apply function using swifter
        df[[code_col, desc_col]] = df[original_col].swifter.apply(extract_code_and_description)
```

```python
import pandas as pd
import swifter

# Define the flattening logic for a single row
def flatten_violations(row):
    flattened = []
    for i in range(1, 26):
        desc = row.get(f'Violation Description - {i}')
        code = row.get(f'Violation Code - {i}')
        memo = row.get(f'Violation Memo - {i}')
        if pd.notna(desc):
            flattened.append({
                'Restaurant Name': row['Restaurant Name'],
                'Inspection Type': row['Inspection Type'],
                'Inspection Date': row['Inspection Date'],
                'Inspection Score': row['Inspection Score'],
                'Street Address': row['Street Address'],
                'Zip Code': row['Zip Code'],
                'Lat Long Location': row['Lat Long Location'],  # ✅ Added here
                'Violation Description': desc,
                'Violation Code': code,
                'Violation Memo': memo
            })
    return flattened

# Apply flattening function in parallel
flattened_rows_nested = df.swifter.apply(flatten_violations, axis=1)

# Flatten the list of lists into a single list
flattened_rows = [violation for sublist in flattened_rows_nested for violation in sublist]

# Final long-form DataFrame
df_flattened = pd.DataFrame(flattened_rows)
```

```python


# Create 'Result' column based on 'Inspection Score'
def get_results(score):
    if score >= 90:
        return 'Pass'
    elif 75 <= score < 90:
        return 'Pass w/ Conditions'
    else:
        return 'Fail'

df_flattened['Results'] = df_flattened['Inspection Score'].apply(get_results)

# Create 'Risk' column based on 'Inspection Score'
def get_risk(score):
    if score >= 90:
        return 'Low'
    elif 75 <= score < 90:
        return 'Medium'
    else:
        return 'High'

df_flattened['Risk'] = df_flattened['Inspection Score'].apply(get_risk)
```

```python
import numpy as np
# Extract numbers in parentheses for 'Latitude' and 'Longitude'
def extract_lat_lon(location_str):
    match = re.search(r'\(([^)]+)\)', str(location_str))
    if match:
        parts = match.group(1).split(', ')
        if len(parts) == 2:
            return parts[0], parts[1]
    return np.nan, np.nan

df_flattened[['Latitude', 'Longitude']] = df_flattened['Lat Long Location'].apply(lambda x: pd.Series(extract_lat_lon(x)))
```

```python
df_flattened['Latitude'] = df_flattened['Latitude'].astype(float).round(5)
df_flattened['Longitude'] = df_flattened['Longitude'].astype(float).round(5)
```
